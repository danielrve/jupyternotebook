## **In Machine Learnig**

#### We have the input and output and try to define the algorithm with models of learning.

#### The inputs and outputs that we have works for training our model for predict futures outputs.

#### The **neurons** or layers connect each other with a **conection**, Each of these connections has an assigned **"WEIGHT"**, which is actually a numerical value that represents the **importance** of the connection between the neurons.

#### Each layer except the input layer has a **"BIAS"** that is also represented in a numerical value.

#### **Loss** is the penalty for a bad prediction. That is, loss is a number indicating how bad the model’s prediction was on a single example.

####If the model’s prediction is perfect, the loss is **zero**; otherwise, the loss is greater. The goal of training a model is to find a **set of weights and biases**  that have **low loss**, on average, across all examples.

## **Let's Started**

This is a example training a model for convert Celsius to Fahrenheit - C to F

```python
# AI Library
import tensorflow as tf
# Array Number handler
import numpy as np
```

```python
celsius = np.array([-40, -10, 0, 8, 15, 22, 38], dtype=float)       # Input Layer
fahrenheit = np.array([-40, 14, 32, 46, 59, 72, 100], dtype=float)  # Output Layer -- Results
```

```python
# Keras -- Neuronal Network Handler

# Define Input Layer with (N) neuron
layer = tf.keras.layers.Dense(units=1)
input = tf.keras.Input(shape=(1,))
# Choice NN Model -- Sequential
model = tf.keras.Sequential([input,layer])
```

```python
model.compile(
# Adam let define weights and biases
    optimizer=tf.keras.optimizers.Adam(0.1),
    loss='mean_squared_error'
)
```

```python
# Begining the Trainning
print("The trainning it's beging...\n")

import time as t

# Cicle for optimization -- how many times try
cicle=1000

start = t.time()

history= model.fit(celsius,fahrenheit,epochs=cicle, verbose=False)

stop = t.time()

print("Training it's over in ", stop-start, "seconds" )
```

## **Before start predicting.. check loss function**
#### That tell you how bad is the Network in each cicle

```python
import matplotlib.pyplot as plt
plt.xlabel("# Cicles")
plt.ylabel("Loss Magnitude")
plt.plot(history.history["loss"])
```

#### **Graph show since 600 cicles the model stop learning**

```python
import pandas as pd

df = pd.DataFrame(history.history["loss"])

df.columns = ['Loss']

df
```

```python
import math as mt
import numpy as np

print("Let's make a prediction!\n")

print("What C temperature would you like to calculate?")
celsius_temp_input = 32

celsius_temp = float(celsius_temp_input)
# Convert the input to a NumPy array
result = model.predict(np.array([celsius_temp]))

result = int(mt.ceil(result.item()))


print(f"\n The result is {result} Fahrenheit")
```

```python
# Check Model Vars

print("Internals Vars from the trained model\n")
print(layer.get_weights())
```

```python

mesure = layer.get_weights()
weight = mesure[0][0]
bais = mesure[1][0]

weight = float(weight.item())
bais = int(mt.ceil(bais.item()))

print(f"The weight is : {weight:.2f}" )
print(f"The bais is : {bais}" )
```
