## **Notebook for ML - Learning how calculate Celsius to Fahrenheit**

nns_bias_2 (1).png

```python
# AI Library
import tensorflow as tf     
# Array Number handler
import numpy as np         
```

```python
celsius = np.array([-40, -10, 0, 8, 15, 22, 38], dtype=float)       # Input Layer
fahrenheit = np.array([-40, 14, 32, 46, 59, 72, 100], dtype=float)  # Output Layer -- Results
```

```python
# Keras -- Neuronal Network Handler

# Define Input Layer with (N) neuron
layer = tf.keras.layers.Dense(units=1, input_shape=[1])  
# NN Model -- Sequential
model = tf.keras.Sequential([layer]) 
```

```python
model.compile(
# Adam let define weights and biases
    optimizer=tf.keras.optimizers.Adam(0.1),
    loss='mean_squared_error'
)
```

```python
# Begining the Trainning
print("the Trainning it's beging...")

# Cicle for optimization -- how many times try
cicle=1000

history= model.fit(celsius,fahrenheit,epochs=cicle, verbose=False)

print("Training it's over")
```

```python
import matplotlib.pyplot as plt
plt.xlabel("# Cicles")
plt.ylabel("Loss Magnitud")
plt.plot(history.history["loss"])
```

#### Graph show since 600 cicles the model stop learning

```python
print("Let's make a prediction!")
result = model.predict([0])
import math as mt
print("The result is "+ str(int(mt.ceil(result))) + " fahrenheit")
```
